The dataset is comprised of several thousand historical text based articles published by BBC News. ## __Import required libraries and initialize spark session__

```python
# Pyspark libraries
import pyspark
from pyspark.sql.functions import col, lower, rand
from pyspark.ml.feature import RegexTokenizer, StopWordsRemover, CountVectorizer, OneHotEncoder, StringIndexer, VectorAssembler, HashingTF, IDF, Word2Vec
from pyspark.ml import Pipeline
from pyspark.ml.classification import LogisticRegression, NaiveBayes, DecisionTreeClassifier, RandomForestClassifier 
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder, CrossValidatorModel

# Other libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
np.random.seed(25)
```

## __Load data, apply transformations, split data into train/test sets__

```python
# Define function to read in data and create spark dataframe, 
def load_data(data_path, data_format, inferHeader_bool=True, inferSchema_bool=True, inferTimestamp_bool=True):
    df = spark.read.format(data_format)\
              .option('header',inferHeader_bool)\
              .option('inferSchema', inferSchema_bool)\
              .option('timestamp', inferTimestamp_bool)\
              .load(data_path)
    return(df)
```

```python
# Apply transformations to data and print summary information 
df = load_data('BBC_News_Train.csv', 'csv')


df = df.select(lower(col('Text')),lower(col('Category')))\
                .withColumnRenamed('lower(Category)','Category')\
                .withColumnRenamed('lower(Text)', 'Description')
df.cache()
print('Dataframe Schema:')
print(df.printSchema())

print('\nDataframe first 10 rows:')
print(df.show(10))

training, test = df.randomSplit([0.7,0.3], seed=48)
print('\nDataset Count: ', df.count())
print("Train Dataset Count:", training.count())
print("Test Dataset Count:", test.count())
```

## __Define NLP functions to be used in the model__
__In general there are 4 principal steps needed for any NLP Model (first four points below)__

* __Tokenization__:<br>
This is the process of dividing up text into individual words and removing any unnecessary characters or words.

* __Stopword removal / data cleaning__:<br>
We must also remove any stopwords including, 'the', 'to', 'this', 'are', 'a' etc. since including them in the model adds little value and increases computation time.

* __Stemming and lemmatization__:<br>
Stemming and lemmatization are both processes that reduce the length of each word. Stemming attempts to reduce the following, 'liked', 'likes', 'likely', 'liking', to the root word 'like'. While stemming operates on a single word only, lemmatization looks at the vocabulary and context of a word. For example lemmatization would be able to find the lemma for the word 'better' to be 'good', while stemming would miss this.

* __Converting categorical labels into numeric format__:<br>
We can use the 'Bag of Words' method to encode the string text data in a numerical format (binary value).
Count vectorization improves on BoW by not only looking at whether a word appears in the text, but how many times it appears.

* __TF-IDF__<br>
TF-IDF which means Term Frequency and Inverse Document Frequency is a method used to calculate how relevant a term is in a given document. A more generic definition for idf would be a weight that upweights less frequent terms and downweights more frequent terms.

* __Word2Vec__<br>
Word2Vec is a class of algorithms that generate vectorized representations of words, based on their linguistic context. These word vectors are obtained by training a shallow neural network (single hidden layer) on individual words in a text, and given surrounding words as the label to predict. By training a neural network on the data, we update the weights of the hidden layer, which we extract at the end as the word vectors.

```python
# Tokenizer 
regex_tokenizer = RegexTokenizer(pattern='\\W').setInputCol("Description").setOutputCol("tokens")

# StopwordsRemover
stopwords_remover = StopWordsRemover().setInputCol('tokens').setOutputCol('filtered_words')
                    
# CountVectorizer / Bag of Words
count_vectors = CountVectorizer(vocabSize=10000, minDF=5).setInputCol("filtered_words").setOutputCol("features")

# TF-IDF 
hashingTf = HashingTF(numFeatures=1000).setInputCol("filtered_words").setOutputCol("raw_features")
            
# MinDocFreq
idf = IDF(minDocFreq=5).setInputCol("raw_features").setOutputCol("features")

# Word2Vec
word2Vec = Word2Vec(vectorSize=1000, minCount=0).setInputCol("filtered_words").setOutputCol("features")

# StringIndexer / Encode categorical variable into label using 
label_string_idx = StringIndexer().setInputCol("Category").setOutputCol("label")
```

## __Define function to evaluate metrics__

```python
def evaluate_metrics(predictions, metrics, print_bool):
    if 'all' in metrics:
        metrics = ['accuracy', 'f1', 'precisionByLabel', 'recallByLabel']
      
    metrics_eval = []
    for metric in metrics:
        evaluator = MulticlassClassificationEvaluator(
        labelCol="label", predictionCol="prediction", metricName=metric)
        metric_eval = evaluator.evaluate(predictions)
        metrics_eval.append(metric_eval)

        if print_bool==True:
            print('%s: %s' %(metric, metric_eval))
    return(metrics_eval)
```

## __Define classifiers (Discriminative and Generative models)__


In Short, the generative approach is to learn the joint distribution P(x,y), whereas the discriminative approach is to learn the conditional distribution P(y|x). In general, the generative approach is harder since P(x,y) = P(y|x) P(x). If the data is x and the label is y, then the posterior (predictive) is P(y|x).

Generative:
* Naive Bayes

Discriminative:
* Logistic Regression
* Decision Trees 
* Random Forest

```python
# Logistic Regression
lr = LogisticRegression(maxIter=20, regParam=0.01, elasticNetParam=0)

# Naive Bayes
nb = NaiveBayes(smoothing=1)

# Decision Tree
dtc = DecisionTreeClassifier(maxDepth = 6)

# Random Forest 
rf = RandomForestClassifier()
```

```python
pipeline_dict = {'lr_cv': [regex_tokenizer, stopwords_remover, count_vectors, label_string_idx, lr],
                'nb_cv': [regex_tokenizer, stopwords_remover, count_vectors, label_string_idx, nb],
                'dtc_cv': [regex_tokenizer, stopwords_remover, count_vectors, label_string_idx, dtc],
                'rf_cv': [regex_tokenizer, stopwords_remover, count_vectors, label_string_idx, rf],
                'lr_idf': [regex_tokenizer, stopwords_remover, hashingTf, idf, label_string_idx, lr],
                'nb_idf': [regex_tokenizer, stopwords_remover, hashingTf, idf, label_string_idx, nb],
                'dtc_idf': [regex_tokenizer, stopwords_remover, hashingTf, idf, label_string_idx, dtc],
                'rf_idf': [regex_tokenizer, stopwords_remover, hashingTf, idf, label_string_idx, rf],
                'lr_w2v': [regex_tokenizer, stopwords_remover, word2Vec, label_string_idx, lr],
                'nb_w2v': [regex_tokenizer, stopwords_remover, word2Vec, label_string_idx, nb]}
```

```python
def run_pipeline(train_data, test_data, stages, metrics, print_bool):

    pipeline = Pipeline().setStages(stages)
    model = pipeline.fit(train_data)
    predictions = model.transform(test_data)

    print('Metrics:')
    evaluated_metrics = evaluate_metrics(predictions, metrics, print_bool)
    
    return(evaluated_metrics)
```

## __Perform cross-validation to tune hyperparameter in models__

K-Fold cross validation is used to optimizing the hyperparameters of the models 

```python
# Linear Regression
param_grid = ParamGridBuilder().addGrid(
    idf.minDocFreq, [1, 10]
    ).addGrid(
    hashingTf.numFeatures, [1000, 10000, 50000]
    ).addGrid(
    lr.regParam, [0.01, 0.05, 0.1, 0.3]
    ).addGrid(
    lr.elasticNetParam, [0.0, 0.3, 0.6, 0.8]
    ).build()

stages_ = pipeline_dict['lr_cv']
pipeline = Pipeline(stages=stages_)

cv = CrossValidator(estimator=pipeline, 
                    estimatorParamMaps=param_grid, 
                    evaluator=MulticlassClassificationEvaluator(
        labelCol="label", predictionCol="prediction", metricName='accuracy'), 
                    numFolds=3)

cvModel = cv.fit(training)

best_param = cvModel.getEstimatorParamMaps()[np.argmax(cvModel.avgMetrics)]
```

```python
# Naive Bayes
param_grid = ParamGridBuilder().addGrid(
    idf.minDocFreq, [1, 10]
    ).addGrid(
    hashingTf.numFeatures, [1000, 10000, 50000]
    ).addGrid(
    nb.smoothing, [1, 2, 3, 4, 5]
    ).build()

stages_ = pipeline_dict['nb_cv']
pipeline = Pipeline(stages=stages_)

cv = CrossValidator(estimator=pipeline, 
                    estimatorParamMaps=param_grid, 
                    evaluator=MulticlassClassificationEvaluator(
        labelCol="label", predictionCol="prediction", metricName='accuracy'), 
                    numFolds=3)

cvModel = cv.fit(training)

best_param = cvModel.getEstimatorParamMaps()[np.argmax(cvModel.avgMetrics)]
```

## __Compare models__


We can also use K-Fold cross validation to compare several model forms. Note, PySparks standard Cross Validation function provides very limited access to the submodels, therefore the below function was required to create.

```python
# Cross validation function from scratch. Needed since pyspark CrossValidation given limited access to submodels
class CrossValidatorScratch(CrossValidator):

    def _fit(self, dataset):
        est = self.getOrDefault(self.estimator)
        epm = self.getOrDefault(self.estimatorParamMaps)
        numModels = len(epm)

        eva = self.getOrDefault(self.evaluator)
        metricName = eva.getMetricName()

        nFolds = self.getOrDefault(self.numFolds)
        seed = self.getOrDefault(self.seed)
        h = 1.0 / nFolds

        randCol = self.uid + "_rand"
        df = dataset.select("*", rand(seed).alias(randCol))
        metrics = [0.0] * numModels

        #Container for metrics 
        metricsReturn = []
        
        for i in range(nFolds):
            foldNum = i + 1

            validateLB = i * h
            validateUB = (i + 1) * h
            condition = (df[randCol] >= validateLB) & (df[randCol] < validateUB)
            validation = df.filter(condition)
            train = df.filter(~condition)

            for j in range(numModels):
                paramMap = epm[j]
                model = est.fit(train, paramMap)
                metric = eva.evaluate(model.transform(validation, paramMap))
                metrics[j] += metric
                metricsReturn.append(metric)
                avgSoFar = metrics[j] / foldNum
 
        if eva.isLargerBetter():
            bestIndex = np.argmax(metrics)
        else:
            bestIndex = np.argmin(metrics)

        bestParams = epm[bestIndex]
        bestModel = est.fit(dataset, bestParams)
        avgMetrics = [m / nFolds for m in metrics]
        bestAvg = avgMetrics[bestIndex]

        return self._copyValues(CrossValidatorModel(bestModel, avgMetrics)), metricsReturn
```

```python
# Function to run CrossValidationScratch with clear inputs
def run_CrossValidatorScratch(data, stages, metricName, kFolds, return_model_bool):

    param_grid = ParamGridBuilder().build()
    pipeline = Pipeline().setStages(stages)
    
    cv = CrossValidatorScratch(estimator=pipeline, 
                    estimatorParamMaps=param_grid, 
                    evaluator=MulticlassClassificationEvaluator(
                    labelCol="label", 
                    predictionCol="prediction", 
                    metricName='accuracy'), 
                    numFolds=kFolds)
    modelCV, metricsCV = cv.fit(data)
    
    if return_model_bool == True:
        return modelCV, metricsCV
    
    else:
        return metricsCV
```

## __Model Comparison Results__

Below we can see Naive Bayes with Count Vectorizer, a reletively simple model, performed the best. Most models performed reletively well with the exception of the random forest and decision trees. Suprisingly Count Vectorizer outperformed TF-IDF - In general, TF-IDF is better than Count Vectorizers because it not only focuses on the frequency of words present in the corpus but also provides the importance of the words. 

With more effort tuning the TF-IDF parameters, the performance should at least rival that of Count Vectorizor.

```python
# Function to plot box-plots for CV results
def plot_cv_results(cv_result_list):

    for i in range(len(cv_result_list)):
        if i == 0:
            metrics_Stack = np.vstack((np.column_stack((len(cv_result_list[i][0]) * [cv_result_list[i][1]], cv_result_list[i][0]))))
        else:       
            metrics_Stack = np.vstack((metrics_Stack, np.column_stack((len(cv_result_list[i][0]) * [cv_result_list[i][1]], cv_result_list[i][0]))))

    metrics_Stack = pd.DataFrame(metrics_Stack, columns = ['model_name', 'accuracy'])                       
    metrics_Stack['model_name'] = metrics_Stack.model_name.astype(str)
    metrics_Stack['accuracy'] = metrics_Stack.accuracy.astype(float)
    
    sns.set(rc={'figure.figsize':(12, 8.5)})
    sns.boxplot(x='model_name', y='accuracy', data=metrics_Stack)
    sns.stripplot(x='model_name', y='accuracy', data=metrics_Stack, size=8, jitter=True, edgecolor='gray', linewidth=2)
```
